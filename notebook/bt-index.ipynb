```python
import bt
import numpy as np
import yfinance as yf
import traceback
import logging
```

```python
# Lấy dữ liệu
data = bt.get('BTC-USD', start='2019-01-01')
```

```python
# Tạo chiến lược
class MAcrossover(bt.Algo):
    def __init__(self, short_window, long_window):
        self.short_window = short_window
        self.long_window = long_window

    def __call__(self, target):
        # Tính toán MA
        short_ma = target.universe.rolling(self.short_window).mean()
        long_ma = target.universe.rolling(self.long_window).mean()

        # Xử lý NaN (nếu cần)
        # short_ma = short_ma.fillna(0)  # Hoặc sử dụng .dropna() để loại bỏ các hàng chứa NaN
        # long_ma = long_ma.fillna(0)
        # Xác định tín hiệu mua/bán
        entries = np.logical_and((short_ma > long_ma), (short_ma.shift(1) <= long_ma.shift(1)))
        exits = np.logical_and((short_ma < long_ma), (short_ma.shift(1) >= long_ma.shift(1)))
        # print("entries", entries)
        # print("exits", exits)
        entries = entries.fillna(False)
        exits = exits.fillna(False)
        
        # Điều chỉnh trọng số danh mục đầu tư
        try: 
            if entries.any():
                print("r111") 
                # target.rebalance(1) 
            elif exits.any():
                print("r222") 
                # target.rebalance(0)
        except Exception as e:
            # print("entries", entries)
            print("entries any", entries.any())
            entries.to_csv('entries_tables.csv')
            logging.error(traceback.format_exc())
            return False

        # print("some here")
        return True
```

```python
# Khởi tạo chiến lược
s = bt.Strategy('MA Crossover', 
                [MAcrossover(20, 120),
                 bt.algos.Rebalance()])
```

```python
# Backtest
test = bt.Backtest(s, data)
res = bt.run(test)
```

```python
# Hiển thị kết quả
res.plot()
res.display()
```
